## Prétraitement des données

```python
# Séparer les caractéristiques (X) et la variable cible (y)
X = data.drop('Prix', axis=1)
y = data['Prix']
```

```python
print(X)
```

```python
print(y)
```

##Gestion des valeurs manquantes


```python
# Obtenir le nombre total de valeurs nulles dans chaque colonne :
data.isnull().sum()
```

```python
# Identifier les colonnes avec des valeurs manquantes
cols_with_nan = data.columns[data.isnull().any()]
cols_with_nan
```

```python
# Séparer les colonnes numériques et catégorielles avec valeurs manquantes
numeric_cols = data[cols_with_nan].select_dtypes(include=['number']).columns
categorical_cols = data[cols_with_nan].select_dtypes(include=['object']).columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Imputer les valeurs manquantes dans les colonnes numériques avec la moyenne
numeric_imputer = SimpleImputer(strategy='mean')
data[numeric_cols] = numeric_imputer.fit_transform(data[numeric_cols])
data[numeric_cols]
```

```python
# Imputer les valeurs manquantes dans les colonnes catégorielles avec la valeur la plus fréquente
categorical_imputer = SimpleImputer(strategy='most_frequent')
data[categorical_cols] = categorical_imputer.fit_transform(data[categorical_cols])
data[categorical_cols]
```

## Mise à échelle (Feature Scaling)






```python
# Mise à l'échelle des variables numériques
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_cols = ['Surf', 'Cham_c', 'Salle_bain', 'Parking', 'Prix_Sqft', 'Prix']
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])
data[numerical_cols]
```

## Encodage des variables catégorielles

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
# Création du ColumnTransformer pour encoder les variables catégorielles
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [3, 4, 7, 8, 9])],
    remainder='passthrough')
```

```python
# Appliquer la transformation et convertir X en tableau numpy
X = ct.fit_transform(data)
X
```

```python
# Convertir la matrice creuse en un tableau dense
X_dense = X.toarray()

# afficher le résultat en tant que DataFrame pour une meilleure lisibilité
data_encoded = pd.DataFrame(X_dense, columns=ct.get_feature_names_out())
print(data_encoded)
```

## Répartition de l'ensemble des données (Entrainement/test)

```python
from sklearn.model_selection import train_test_split
```

```python
# Répartition des données en ensembles d'entraînement (80%) et de test (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Entrainement du modèle

```python
from sklearn.linear_model import LinearRegression
```

```python
# Créer et entraîner le modèle de régression linéaire multiple
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Extrait les noms des colonnes (les variables explicatives) à l'exception de la variable cible
noms_variables = data.columns[data.columns != "Prix"]
noms_variables
```

```python
# Obtenir les estimations des paramètres du modèle
coefficients =regressor.coef_
intercept = regressor.intercept_
```

### Afficher les estimations des paramètres

```python
print(coefficients)
```

```python
print(intercept)
```

```python
# Créer un dictionnaire pour stocker les coefficients avec les noms des variables correspondantes
coefficients_dict = dict(zip(noms_variables, coefficients))

# Trier les coefficients par valeur absolue pour identifier les variables les plus importantes
coefficients_dict_sorted = dict(sorted(coefficients_dict.items(), key=lambda item: abs(item[1]), reverse=True))

# Afficher les coefficients avec leurs noms de variable
for variable, coefficient in coefficients_dict_sorted.items():
    print(f"{variable}, Coefficient : {coefficient}")
```

## Prédiction des résultats de l'ensemble de test

```python
# Prédire les résultats de l'ensemble de données test
y_pred = regressor.predict(X_test)
print(y_pred)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.values.reshape(len(y_test),1)), axis=1))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Calcul du coefficient de détermination (R²)
r_squared = r2_score(y_test, y_pred)
print("R² :", r_squared)
```

Le coefficient de détermination (R²) de 0.9999999996504522 indique une excellente adéquation entre les valeurs prédites par notre modèle et les valeurs réelles de l'ensemble de test. En d'autres termes, environ 99,99999996504522% de la variance dans la variable dépendante est expliquée par les variables indépendantes incluses dans notre modèle.

```python
# Calcul de l'erreur quadratique moyenne (MSE)
mse = mean_squared_error(y_test, y_pred)
print("MSE :", mse)
```

Une MSE de 237495.46295755604 est considérée comme élevée, ce qui suggère que le modèle de régression a une performance relativement faible.
